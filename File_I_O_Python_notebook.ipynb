#### File object methods
```
* open(): opens a file; returns a file object
* close(): closes an open file
* read(): reads from an open file
* readline(): reads a line from an open text file
* readlines(): reads all lines from an open file, and stores each line in a list
* write(): writes a string to a file
* tell(): returns the current position of the file pointer (number of bytes from beginning of file)
* seek(): changes the position of the file pointer

```

| ```rindex()``` | Searches the string for a specified value and returns the last position of where it was found |
| ```rstrip()``` | Returns a right trim version of the string |
| ```split()``` | Splits the string at the specified separator, and returns a list |
| ```splitlines()``` | Splits the string at line breaks and returns a list |
| ```strip()``` | Returns a trimmed version of the string |
| ```upper()``` | Converts a string into upper case |

the use of a few of these will be illustrated in the code cell below

````python
# code cell 6

pangram = "'The quick brown fox jumped over the lazy dog.'"
print(pangram + ' is a pangram.')
print('the letter \'e\' occurs', end=" ")
print(pangram.count('e'), end=" ")
print('times')

print('the letter \'e\' occurs at positions', end=" ")
i1 = pangram.find('e')
i2 = pangram.find('e', i1 + 1)
i3 = pangram.find('e', i2 + 1)
i4 = pangram.rfind('e')
print(str(i1) + ', ' + str(i2) + ', ' + str(i3) + ', and ' + str(i4))
print()

print('i can also find the indices of where the letter \'e\' occurs using index():', end = " ")
j1 = pangram.index('e')
j2 = pangram.index('e', j1 + 1)
j3 = pangram.index('e', j2 + 1)
j4 = pangram.rindex('e')
print(str(j1) + ', ' + str(j2) + ', ' + str(j3) + ', and ' + str(j4))
print()

print('is \'X\' in pangram? ' + str('X' in pangram))
print('\'X\' occurs ' + str(pangram.count('X')) + ' times in pangram')
print('the letter \'X\' occurs at index ' + str(pangram.find('X')) + ', which means "not found"')
print()

pList = pangram.split()
print('pangram as a list: ' + str(pList))
pStr = ' '.join(pList) 
print('pList as a string: ' + pStr)
print()

print('pangram is all ASCII characters: ' + str(pangram.isascii()))
print('pangram is all alphanumeric characters: ' + str(pangram.isalnum()))
print('now replace single quotes with nothing: ' + str(pangram.replace("'", "")))
pangram2 = pangram.replace("'", "")
print('pangram2 = ' + pangram2)
print('now replace period with newline: ' + str(pangram2.replace(".", "\n")))
pangram3 = pangram2.replace(' ', '').strip()
print('now replace spaces in pangram2 with nothing: ' + pangram3)
print('pangram3 is all alphabetic characters: ' + str(pangram3.isalpha()))
print('pangram3 is all alphanumeric characters: ' + str(pangram3.isalnum()))
print()

pangram4 = pangram3.upper()
print(pangram4)
print(pangram4.lower())
````

**binary files**

the bitmap file format: https://medium.com/sysf/bits-to-bitmaps-a-simple-walkthrough-of-bmp-image-format-765dc6857393

file ```bmp_24.bmp``` (200x200 pixel bitmap file, 24 bpp, file size 120,054 bytes)

bmp_24.bmp

use of ```seek(offset, whence)```

```offset```:  number of bytes from ```whence```

```whence```:  0 = beginning of file, 1 = current file pointer position, 2 = end of file

````python
# code cell 9

# 200x200 pixel 3-color bitmap file
# size 120,054 bytes (0x01D4F6 in hex), 24 bits per pixel (bpp)
def main(fName):
    f = open(fName, 'rb')  # open file for read in binary mode
    print('opened file ' + fName + '\n')

    print('current file pointer position: ' + str(f.tell()))
    x = f.read(2)  # read two bytes at current pointer location
    print("first two bytes of file are 'BM' at offset 0: " + str(x) + '\n')

    print('current file pointer position: ' + str(f.tell()))
    x = f.read(4)  # read four bytes at current pointer location
    print('file size in bytes (4 bytes at offset 2): ' + str(x) + '\n')

    f.seek(18, 0)  # move pointer to 18 bytes from start
    print('current file pointer position: ' + str(f.tell()))
    x = f.read(4)  # read four bytes at current pointer location
    print('horiz width (4 bytes at offset 18): ' + str(x) + '\n')

    f.seek(0, 1)   # move pointer 0 bytes from current location
    print('current file pointer position: ' + str(f.tell()))
    x = f.read(4)  # read four bytes at current pointer location
    print('vert height (4 bytes at offset 22): ' + str(x) + '\n')

    f.seek(2, 1)   # move pointer 2 bytes from current location
    print('current file pointer position: ' + str(f.tell()))
    x = f.read(2)  # read two bytes at current pointer location
    print('bits per pixel (2 bytes at offset 28): ' + str(x))

    f.close()

# replace with your own pathname to bmp_24.bmp
fName = "C:\\Users\\EricKawamoto\\Dropbox\\Documents\\teaching\\BATS\\COS 200 A block\\bmp_24.bmp" 
main(fName)
````
